```python
def to_one_hot(word_dico):
    import keras
    dic={}
    a=[]
    for i in range(0,len(word_dico)):
        a.append(i)
    one = keras.utils.to_categorical(a)
    for i in word_dico:
        dic[word_dico[int(i)]] = one[int(i)]
    return dic
```

```python
word_enco = to_one_hot(word_dic)
print(word_enco)
```
